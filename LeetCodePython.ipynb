```python
class Solution():
 def twoValues(self, num, target):
  for i in range(len(num)):
    for j in range(i+1, len(num)):
        if num[i]+num[j]==target:
            return[i,j]
sl=Solution()
num=[1,7,8,9,5,48]
target=15
print(sl.twoValues(num, target))
            
```

```python
class Fun():
    def sum(self,it):
        for x in it:
          print(x)
sn=Fun()
it=[1,2,3,4,5,6]
sn.sum(it)
```

```python
class Fun():
    def dm(self,add):
        for x in range(len(add)):
            print(add[x])
sc=Fun()
add=[1,8,9,4,5]
sc.dm(add)
```

```python
class Solution:
    def isPallendrom(self, x):
        if str(x)==str(x)[::-1]:
            return True
        else:
            return False
ss=Solution()
print(ss.isPallendrom(125))
print(ss.isPallendrom(151))
```

```python
#By using for loop
class Solution:
    def isd(self, x):
        y=""
        for i in str(x):
            y=i+y
        if str(x)==y:
            return True
        else:
            return False

se=Solution()
print(se.isd(125))
print(se.isd(151))            
                
```

```python
class Solu:
    def ids(self, x):
        y=""
        for i in x:
            y=i+y
        if x==y:
            return True
        else:
            return False
sw=Solu()
print(sw.ids("MOM"))
x="Eleven"
x=x.lower()
print(x)
print(sw.ids("ELEVEN"))
```
