```python
class Solution():
 def twoValues(self, num, target):
  for i in range(len(num)):
    for j in range(i+1, len(num)):
        if num[i]+num[j]==target:
            return[i,j]
sl=Solution()
num=[1,7,8,9,5,48]
target=15
print(sl.twoValues(num, target))
            
```

```python
class Solution:
    def isPallendrom(self, x):
        if str(x)==str(x)[::-1]:
            return True
        else:
            return False
ss=Solution()
print(ss.isPallendrom(125))
print(ss.isPallendrom(151))
```

```python
class Sol:
    def tre(self, nums):
        n=[]
        for i in range(len(nums)):
            for j in range(len(nums)):
                if i!=j and nums[i]==nums[j]:
                    break
            else:
                n.append(nums[i])
        return n
cl=Sol()
nums="theamongths"
print(len(nums))
print(cl.tre(nums))
```

```python
class Sol:
    def tre(self, nums):
        n=[]
        for i in range(len(nums)):
            if nums[i] not in n:
                n.append(nums[i])
        return n
cl=Sol()
nums="theths"
print(len(nums))
print(cl.tre(nums))
```

```python
class Sol:
    def tre(self, nums):
        nums=set(nums)
        nums=sorted(nums)
        return nums
cl=Sol()
nums="theths"
print(len(nums))
print(cl.tre(nums))
```

```python
class Solution(object):
    def threeSum(self, nums):
        re=set()
        for i in range(len(nums)):
            for j in range(len(nums)):
                for k in range(len(nums)):
                    if nums[i]+nums[j]+nums[k]==0:
                        re.add(tuple(sorted([nums[i],nums[j],nums[k]])))
        return list(re)
cd=Solution()
nums=[-1,0,1,2,-1,-4]
print(cd.threeSum(nums))
```
