```python
class Twoesu:
     def eqro(self,ead):
         roman={
             'I': 1,
             'V': 5,
             'X': 10,
             'L': 50,
             'C': 100,
             'D': 500,
             'M':1000,
         }
         e=0
         for i in range(len(ead)):
             if i+1 <len(ead) and roman[ead[i]]<roman[ead[i+1]]:
                 e=e-roman[ead[i]]
             else:
                 e=e+roman[ead[i]]
         return e
sd=Twoesu()
ead="IX"
print(sd.eqro(ead))
```

```python
class Solution(object):
    def romanToInt(self, sod):
        word={}
        for i in sod:
            if i in word:
                word[i]=word[i]+1
            else:
                word[i]=1
        re=[]
        for i in word:
            if word[i]==1:
                re.append(i)
        return re
cd=Solution()
sod="Tsswt"
print(cd.romanToInt(sod))
```
